### Option 2: Read PDF (I chose this as it is more "Table" focused)

```python
#import pdf reader package --tabula-py
from tabula import read_pdf
```

```python
#read pages from pdf
tables = read_pdf('CertStmtCMLT06AMC12210.pdf',pages='all')
```

```python
#count all tables in file
len(tables)
```

```python
#view 4th table in pdf file
df2=tables[4]
df2
```

```python
#get row that contains 'Total Principal Funds Available'
df2 = df2[df2['Unnamed: 0'].str.contains('Total Principal Funds Available')]
```

```python
#view dataframe
df2
```

```python
#create new df that only takes required columns
df3 = df2.loc[:, ['Unnamed: 0', 'Unnamed: 2']] # This does not make a copy (better).
```

```python
#view dataframe
df3
```

```python
#Rename and clean column names
df3.columns=['Description','Amount']
#view final df
df3
```
